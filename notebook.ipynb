```python
import pandas as pd
import os
from datetime import datetime
```

```python
# Load the Excel file
df = pd.read_excel('Superstore Dataset.xlsx')

# Display the first 5 rows of the dataframe
df.head()
```

```python
df.info()
```

```python
# Get the current date
current_date = datetime.now().date()
```

```python
# Filter the dataframe to include only rows with 'Order Date' up to the current date
df_filtered = df[df['Order Date'].dt.date <= current_date]
```

```python
# Create a 'YearMonth' column for grouping
df_filtered['YearMonth'] = df_filtered['Order Date'].dt.to_period('M')
df_filtered['Year'] = df_filtered['Order Date'].dt.year

# Group by 'YearMonth'
groups = df_filtered.groupby('YearMonth')

df_filtered.head()
```

```python
# Create a new folder for the CSV files
output_folder = 'MonthlyData'
os.makedirs(output_folder, exist_ok=True)
```

```python
# Save each group to a separate CSV file in the appropriate year folder
for name, group in groups:
    # Extract year from the group name
    year = str(name.year)
    # Create year folder if it doesn't exist
    year_folder = os.path.join(output_folder, year)
    os.makedirs(year_folder, exist_ok=True)
    # Format the filename as 'Superstore_YYYY-MM.csv' inside the year folder
    filename = os.path.join(year_folder, f'Superstore_{name}.csv')
    # Save the group to a CSV file
    group.to_csv(filename, index=False)
```
